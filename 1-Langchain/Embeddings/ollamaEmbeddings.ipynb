## Ollama

Ollama supports embedding models, making it possible to build RAG applications that combine text prompts with existing documents or other data

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
 
    OllamaEmbeddings(model = 'gemma:2b') ##by default it uses llama2
)
```

```python
embeddings
```

```python
r1 =embeddings.embed_documents(
    ["ys jagan mohan reddy is the chief minister of Andhra Pradesh",
     "I wish he would be the chief minister of Andhra Pradesh forever"]

)

r1
```

```python
len(r1)
```

```python
query_result = embeddings.embed_query("who is the chief minister of Andhra Pradesh?")
```

```python
len(query_result)
```

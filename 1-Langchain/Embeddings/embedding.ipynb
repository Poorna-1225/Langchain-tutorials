we are converting the text into vectors

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings= OpenAIEmbeddings(model = 'text-embedding-3-large')
```

```python
text = "this is a tutorial on openai embedding"
query_result = embeddings.embed_query(text)
```

```python
embeddings_1024= OpenAIEmbeddings(model = 'text-embedding-3-large', dimensions=1024)
```

```python
text = "this is a tutorial on openai embedding"
query_result_1024 = embeddings_1024.embed_query(text)
```

```python
len(query_result_1024)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader = TextLoader('speech.txt')

documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(separators= '\n',chunk_size =500, chunk_overlap =50)

final_documents = text_splitter.split_documents(documents)

final_documents
```

```python
## converting data chunks into vector embeddings and storing in vector db chroma db

from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embeddings_1024)
db
```

```python
#querying the vector db

query  = "who is the father of Jagan MOhan Reddy?"
retrived_results = db.similarity_search(query)

print(retrived_results)
```
